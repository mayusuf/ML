```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Black_footed_Albatross.jpg',cv2.IMREAD_GRAYSCALE)

plt.imshow(img,cmap='gray',interpolation='bicubic') # convert image in gray scale
plt.plot([100,50],[20,150],'c',linewidth=3) # draw a line over the image
plt.show()
```
